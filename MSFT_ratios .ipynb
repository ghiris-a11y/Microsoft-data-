```python
import pandas as pd
```

```python
income = pd.read_csv('msft_income_statement.csv', index_col=0)
bs = pd.read_csv('msft_balance_sheet.csv', index_col=0)
```

```python
ratios = pd.DataFrame(index=income.index)
```

```python

ratios["Gross Margin"] = income["Gross Profit"] / income["Total Revenue"]
ratios["Operating Margin"] = income["Operating Income"] / income["Total Revenue"]
ratios["Net Margin"] = income["Net Income"] / income["Total Revenue"]
ratios["Current Ratio"] = bs["Current Assets"] / bs["Current Liabilities"]
ratios["Debt to Equity"] = bs["Total Debt"] / bs["Stockholders Equity"]
ratios["Asset Turnover"] = income["Total Revenue"] / bs["Total Assets"]
```

```python

ratios.to_csv("msft_ratios(1).csv")
ratios
```
