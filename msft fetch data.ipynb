```python

ticker = 'msft' 
start_date = '2021-01-01'
end_date = '2025-01-01'

msft_data = yf.download(ticker, start=start_date, end=end_date)

print(msft_data.head())
```

```python
msft_data.to_csv('msft_stock_prices.csv')
```

```python
msft_ticker = yf.Ticker('MSFT')
```

```python
income = msft_ticker.financials.T
balance = msft_ticker.balance_sheet.T
cashflow = msft_ticker.cashflow.T
```

```python
print("Income Statement (Transposed):\n", income.head())
```

```python
income.to_csv('msft_income_statement.csv')
balance.to_csv('msft_balance_sheet.csv')
cashflow.to_csv('msft_cashflow.csv')
```
