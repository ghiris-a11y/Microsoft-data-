```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python

def load_csv(path):
    return pd.read_csv(path, index_col=0)
```

```python
def plot_ratios(ratios_df):
    ratios_df.plot(kind='bar', figsize=(12, 6))
    plt.title('Financial Ratios Over Time')
    plt.xlabel('Year')
    plt.ylabel('Ratio Value')
    plt.tight_layout()
    plt.show()
```

```python
ratios = load_csv('msft_ratios(1).csv')
plot_ratios(ratios)
```
